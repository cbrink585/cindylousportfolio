```python
#Step 1 rename column
df.rename(columns = {'name':'Book'}, inplace = True)
```

```python
df
```

```python
#Step 2 Rename column
```

```python
df.rename(columns = {'chapters':'Chapters'}, inplace = True)
```

```python
df
```

```python
#Step 3 remove unneeded column
```

```python
df.drop('shortname', inplace= True, axis=1)
```

```python
df
```

```python
#Step 4 Reset index
```

```python
df = df.reset_index(drop=True)
```

```python
df
```

```python
#Step 5 Remove quotes around the chapters.
```

```python
df.applymap(lambda x: x.replace('', ' '))#Trying to replace the quotes with spaces.  Doesn't work because they are integers not strings.
```

```python
#Step 5 Remove quotes around  chapters
```

```python
df['chapter_verses'] = df['chapter_verses'].astype(str).replace("[']", "", regex=True)
```

```python
df
```

```python
#Step 6 Rename column.
```

```python
df.rename(columns = {'chapter_verses':'Chapters_Verses'}, inplace = True)
```
